```python
from datetime import datetime, timedelta

def get_past_time(hours=0, minutes=0):
    # Get the current date and time
    current_time = datetime.now()
    days=2

    # Subtract the given hours and minutes
    past_time = current_time - timedelta(days=days,hours=hours, minutes=minutes)

    # Format the date and time to show up to minutes
    formatted_time = past_time.strftime('%Y-%m-%d, %H:%M')

    return formatted_time

# Example usage:
hours_ago = 1
minutes_ago =42

past_time = get_past_time(hours=hours_ago, minutes=minutes_ago)

print(f"{hours_ago} hours and {minutes_ago} minutes ago, the date and time was: {past_time}")
```
